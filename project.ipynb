### let's dieve into ML model

#### let's predict the Risk Level

```python
# selecting freatures for the model

features = [
    'attendance_percentage', 'study_hours_per_day', 
    'previous_semester_gpa', 'current_cgpa', 
    'sleep_hours', 'social_media_hours'
]
X = data[features]
y = data['risk_level'] # the target variable
```

##### Encoding the categoric target variable ( means that to convert it into numeric values like 0,1,2...) bcz this is a rank wise variable(low,medium, high)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

#### Now, let's split the data into train and Test 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

#### Now, let's train the model on Train Data

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)
```

#### Now, let's check the accuracy of the model

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(X_test)
print(f"Accuracy Score: {accuracy_score(y_test, predictions)}")
```

```python
#   let's check whether both gives the same accuracy?
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

print(f"Train Accuracy: {accuracy_score(y_train, train_pred)}")
print(f"Test Accuracy: {accuracy_score(y_test, test_pred)}")
```

```python
# saving the model for the app UI
import joblib

joblib.dump(model, 'student_risk_model.pkl')
joblib.dump(le, 'label_encoder.pkl')
print("\nModel saved as 'student_risk_model.pkl'")
```

```python
# let's check the new data on the model
def predict_student_risk(attendance, study, prev_gpa, curr_cgpa, sleep, social):
    # Model load karein
    m = joblib.load('student_risk_model.pkl')
    l = joblib.load('label_encoder.pkl')
    
    # Input data ko format karein
    student_data = [[attendance, study, prev_gpa, curr_cgpa, sleep, social]]
    
    # Prediction
    res = m.predict(student_data)
    risk = l.inverse_transform(res) # Number se wapis "High/Low Risk" mein badalna
    return risk[0]

# --- EXAMPLE TEST ---
print("\n--- Testing with New Data ---")
test_risk = predict_student_risk(85, 6, 3.4, 3.5, 7, 2)
print(f"Predicted Risk Level: {test_risk}")
```
